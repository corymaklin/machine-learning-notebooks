```python
def calculate_score(u, i):
    # Check whether the item is in the training dataset
    if i not in ratings_matrix_df.columns:
        return 2.5
    if ratings_matrix_df[i].isna().all():
        return 2.5
    print(f"{u},{i}")

    avg_user_rating = ratings_matrix_df.T.mean()[u]
    similarities = similarity_matrix_df[u].drop(labels=u)
    normalized_ratings = normalized_ratings_matrix_df[i].drop(index=u)

    # Drop users that haven't rated the item
    similarities.drop(index=normalized_ratings[normalized_ratings.isnull()].index, inplace=True)
    normalized_ratings.dropna(inplace=True)

    total_score = 0
    total_weight = 0
    for v in similarities.index:
        if not pd.isna(similarities[v]) and not pd.isna(normalized_ratings[v]):
            total_score += normalized_ratings[v] * similarities[v]
            total_weight += abs(similarities[v])
    if total_weight == 0:
        return 2.5
    return avg_user_rating + total_score / total_weight
```
