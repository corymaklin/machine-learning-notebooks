```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
import seaborn as sns
sns.set()
```

```python
np.random.seed(42)
```

```python
a, b = 10, 10
n = 100
h = 59
thetas = np.linspace(0, 1, 200)
```

```python
prior = st.beta(a, b).pdf(thetas)
likelihood = st.binom(n, thetas).pmf(h)
post = prior * likelihood
post /= (post.sum() / len(thetas))
```

```python
plt.figure(figsize=(12, 9))
plt.plot(thetas, prior, label='Prior', c='blue')
plt.plot(thetas, n*likelihood, label='Likelihood', c='green')
plt.plot(thetas, post, label='Posterior', c='red')
plt.xlim([0, 1])
plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel('PDF', fontsize=16)
plt.legend();
```

```python
def f(x):
    return x**2 + 4*x*np.sin(x)

def intf(x): 
    return x**3/3.0 + 4.0*np.sin(x) - 4.0*x*np.cos(x)
```

```python
x = np.linspace(2, 3, 100)
plt.plot(x, f(x), 'r')
```

```python
a = 2;    
b = 3; 
N = 10000

X = np.random.uniform(low=a, high=b, size=N)
Y = f(X)

Imc = (b-a)*np.sum(Y)/N;

exactval = intf(b)-intf(a)
```

```python
print('Monte Carlo estimation: {}'.format(Imc))
print('Exact number: {}'.format(intf(b)-intf(a)))
```
