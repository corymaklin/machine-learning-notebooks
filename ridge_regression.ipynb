```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge
```

```python
X, y, coefficients = make_regression(
    n_samples=50,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=5,
    coef=True,
    random_state=1
)
```

```python
alpha = 1
```

```python
n, m = X.shape
I = np.identity(m)
w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X) + alpha*I), X.T), y)
```

```python
w
```

```python
coefficients
```

```python
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
rr = Ridge(alpha=1)

rr.fit(X, y)

w = rr.coef_
```

```python
w
```

```python
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
rr = Ridge(alpha=10)

rr.fit(X, y)

w = rr.coef_[0]

plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
rr = Ridge(alpha=100)

rr.fit(X, y)

w = rr.coef_[0]

plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```
