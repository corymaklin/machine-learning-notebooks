```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
T = np.array([[0.9, 0.1],[0.5, 0.5]])

p = np.random.uniform(low=0, high=1, size=2)
p = p/np.sum(p)
q=np.zeros((100,2))

for i in np.arange(0,100):
    q[i, :] = np.dot(p,np.linalg.matrix_power(T, i))

plt.plot(q)
plt.xlabel('i')
plt.legend(('S', 'R'))
```
